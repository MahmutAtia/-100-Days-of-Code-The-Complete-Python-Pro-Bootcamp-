```python
import time
def run_time(func):
    def wrapper_func():
        before = time.time()
        func()
        after= time.time()
        print(after - before)
    return wrapper_func
```

```python

def read_book(path):
    with open(path , "r") as f:
        text = f.read()
        text= text.replace("\n", "").replace("\r", "")
    return text
```

```python
def book_stat(count_dic:dict):
    unique_word = len(count_dic)
    num_word = sum(count_dic.values())
    return (unique_word,num_word)
```

```python
#loop over books
import os
import pandas as pd
import matplotlib.pyplot as plt
book_dir = r"C:/Users/user/Documents/books"
pandas_dic = {"language":[],
"auther": [],

                                    "book": [],
                                      "unique words": [],
                                      "number of words" : []
                        }
for language in  os.listdir(book_dir):
    for lang in os.listdir(book_dir +'/'+ language):
        try:
            for auther in os.listdir(book_dir +'/'+ language + "/" + lang):
                try:
                    for title in os.listdir(book_dir +'/'+ language + "/" + lang+ "/"+ auther):
                        text = read_book(book_dir +'/'+ language + "/" + lang+ "/"+ auther+"/"+title)
                        unique , num_word =book_stat(count_word(text))
                        pandas_dic["language"].append(lang)
                        pandas_dic["auther"].append(auther.title())
                        pandas_dic["book"].append(title.replace(".txt", ""))
                        pandas_dic["unique words"].append(unique)
                        pandas_dic["number of words"].append(num_word)
                        
                      
                except:
                        pass 
        except:
              pass

df = pd.DataFrame(pandas_dic)       
df
```

```python
plt.plot(df["number of words"], df["unique words"], "ro")
plt.xlabel("book names")
plt.ylabel("number of words")
```
